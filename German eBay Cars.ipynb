## Cleaning Columns
Now we are going to clean up the columns of the dataframe to make it more python friendly.

I looped through the column attribute which is an array containing all the columns. 

I used regex to substitute the strings which followed the pattern, a capital character followed by a small case character to be replaced by the output of the function lambda which replaced the original part of the string that matched the pattern and replaced it with an underscore along with the match case and applied `.group(0)` which is the original letter that is replaced.

Then I lower cased the output of the previous function which I reassigned back to `var_name` which is now fully clean.

I finally appended this to the clean list which I assigned back to the column attribute.

```python
autos.columns
```

```python
#cleaning the column names
clean_list = []
for i in autos.columns:
    var_name = re.sub('[A-Z][a-z]',lambda x:"_"+x.group(0),i)
    '''Using regex sub function to find capital letters in the string and 
       replace it with a _ along with original letter'''
    var_name = var_name.lower()
    clean_list.append(var_name)
#assigning the list to become the new columns for the dataframe
autos.columns = clean_list
```

```python
#Average prices split up by brand
mean_dict = {}
for i in autos_3['brand'].unique():
    average = autos_3.loc[autos_3['brand']==i,'price_usd'].mean()
    mean_dict[i] = average
```

```python
list_1 = list(mean_dict.items())
mean_list = []
for i in list_1:
    value = i[1]
    key = i[0]
    mean_list.append([value, key])
```

```python
mean_list.sort(reverse=True)
mean_list
```

On the eBay marketplace it seems porsche's on average are listed for the largest price. It doubles the average price for a land rover.  This is probably due to the face porche only making high performance sport vehicles, whilst land rover does produce luxury SUVs, they are generally sold at a much lower RRP compared to a brand new porsche. 

What I find surprising is that mini coopers are on average more costly in the german market compared to higher tier german vehicles such as bmw, audi and mercedes. This is most likely due to the popularity of german vehicles in germany thus there is an abundance in these brand vehicles however the opposite could be said for the british vehicles land rover, jaguar and mini. 

Sonstige_autos is the third most expensive option on eBay Germany. This other category is most likely for cars which have modifications done to them for example Brabus or Volvo's polestar etc. Since these cars would have expensive additions to them, they would most likely cost more than a stock version of the cars itself.

```python
#Average mileage split up by brand
mean_mileage_dict = {}
for i in autos_3['brand'].unique():
    average_km = autos_3.loc[autos_3['brand']==i,'odometer_km'].mean()
    mean_mileage_dict[i] = average_km
```

```python
#Convert dictionaries to series using the constructor
bmp_series = pd.Series(mean_dict, dtype='float64')
bmo_series = pd.Series(mean_mileage_dict, dtype='float64')

brand_averages = pd.DataFrame(bmp_series, columns = ['mean_price'])
brand_averages['mean_mileage'] = bmo_series
```

```python
brand_averages.sort_values(by='mean_price', axis = 0, ascending = False)
```

The mileage sheds some light into why the german brands on average got listed for less than a mini. The german brands on average got listed with an average mileage of around 30000 higher.
